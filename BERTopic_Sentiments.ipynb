```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
analyzer = SentimentIntensityAnalyzer()
df['sentiment'] = df['body'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
# 토픽별 평균 감정 점수 계산
topic_sentiment = df.groupby('topic')['sentiment'].mean()

# 확인
topic_sentiment
```

```python
# 긍정 상위 10개, 부정 상위 10개 선택
top_positive = topic_sentiment[topic_sentiment > 0].sort_values(ascending=False).head(10)
top_negative = topic_sentiment[topic_sentiment < 0].sort_values().head(10)
```

```python
# 긍정 상위 10개와 부정 상위 10개 데이터 준비
positive_df = pd.DataFrame({
    "Topic": [f"Topic {i}" for i in top_positive.index],
    "Sentiment": top_positive.values
})

negative_df = pd.DataFrame({
    "Topic": [f"Topic {i}" for i in top_negative.index],
    "Sentiment": top_negative.values
})

# Subplots 생성
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Positive Sentiments", "Negative Sentiments"),
    column_widths=[0.5, 0.5]
)

# 긍정 막대 그래프 추가
fig.add_trace(
    go.Bar(
        x=positive_df["Topic"],
        y=positive_df["Sentiment"],
        name="Positive",
        marker=dict(color=positive_df["Sentiment"], colorscale="Sunsetdark"),
    ),
    row=1, col=1
)

# 부정 막대 그래프 추가
fig.add_trace(
    go.Bar(
        x=negative_df["Topic"],
        y=negative_df["Sentiment"],
        name="Negative",
        marker=dict(color=negative_df["Sentiment"], colorscale="Sunsetdark"),
    ),
    row=1, col=2
)

# 레이아웃 업데이트
fig.update_layout(
    title_text="Positive and Negative Sentiments by Topic",
    title_font_size=16,
    height=600,
    width=1000,
    showlegend=False
)

# X축 라벨 회전 및 제목 추가
fig.update_xaxes(title_text="Topic", tickangle=45, row=1, col=1)
fig.update_xaxes(title_text="Topic", tickangle=45, row=1, col=2)

# Y축 제목 추가
fig.update_yaxes(title_text="Average Sentiment", row=1, col=1)
fig.update_yaxes(title_text="Average Sentiment", row=1, col=2)

# 그래프 출력
fig.show()
```

```python
# 긍정적인 문서와 해당 토픽
positive_docs = df[df['sentiment'] > 0.2]
print(positive_docs[['body', 'topic']].head())

# 부정적인 문서와 해당 토픽
negative_docs = df[df['sentiment'] < -0.2]
print(negative_docs[['body', 'topic']].head())
```

```python
# 각 토픽의 주요 키워드 가져오기
topic_keywords = {topic: model.get_topic(topic) for topic in topic_sentiment.index}

# 긍정 상위 10개 출력
print("### Top Positive Topics ###\n")
for topic, sentiment in top_positive.items():
    print(f"Topic {topic} - Average Sentiment: {sentiment:.3f}")
    print(f"Keywords: {', '.join([keyword[0] for keyword in topic_keywords[topic]])}\n")

# 부정 상위 10개 출력
print("\n### Top Negative Topics ###\n")
for topic, sentiment in top_negative.items():
    print(f"Topic {topic} - Average Sentiment: {sentiment:.3f}")
    print(f"Keywords: {', '.join([keyword[0] for keyword in topic_keywords[topic]])}\n")
```

```python
# 상위 데이터로 결합
selected_topics = pd.concat([top_positive, top_negative])

# 데이터 준비
heatmap_df = pd.DataFrame({
    'Topic': [f'Topic {i}' for i in selected_topics.index],
    'Sentiment': selected_topics.values
})

# Plotly 히트맵 생성
fig = px.imshow(
    heatmap_df[['Sentiment']].T,  # 히트맵에 사용할 데이터 전치
    labels=dict(x="Topic", y="Metric", color="Sentiment"),
    x=heatmap_df['Topic'],  # X축 레이블
    y=["Sentiment"],  # Y축 레이블
    color_continuous_scale=px.colors.sequential.Sunsetdark,  # 색상 팔레트
    text_auto=True  # 셀에 값 표시
)

# 그래프 레이아웃 설정
fig.update_layout(
    title="Sentiment Heatmap for Selected Topics",
    xaxis_title="Topic",
    yaxis_title="Sentiment",
    height=600,
    width=1000
)

# 그래프 출력
fig.show()
```
